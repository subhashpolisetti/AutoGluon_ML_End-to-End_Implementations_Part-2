# **Fine-Tuning and Cross-Lingual Transfer with AutoGluon**

This notebook demonstrates the process of fine-tuning a sentiment analysis model and evaluating cross-lingual transfer learning using **AutoGluon**. It focuses on leveraging multilingual datasets for fine-tuning and transfer learning tasks.

---

## **Key Objectives**
1. **Fine-Tuning**: Train a German BERT model on a German sentiment dataset for product reviews.
2. **Cross-Lingual Transfer Learning**: Train a model on English data and evaluate its performance on German test data.

---

## **Steps Covered**

### **1. Installation and Setup**
- Install the required version of AutoGluon (`0.8.0`) for text-based machine learning tasks.

### **2. Dataset Preparation**
- Two datasets (English and German) representing product reviews with sentiment labels:
  - **English Data**: Sentences with positive (`1`) or negative (`0`) sentiment.
  - **German Data**: Equivalent sentences translated into German with the same labels.
- Split both datasets into training and testing sets using `train_test_split`.

### **3. Fine-Tuning a German BERT Model**
- Use AutoGluon’s `MultiModalPredictor` to fine-tune a pre-trained BERT model on the German dataset.
- Evaluate the model’s performance on the German test dataset.
- Output accuracy and predictions.

### **4. Cross-Lingual Transfer Learning**
- Train a model on the English dataset.
- Evaluate its performance on the German test dataset to demonstrate cross-lingual transfer learning.
- Compare results with the fine-tuned German model.

---

## **Results**
- Accuracy of the fine-tuned German BERT model on German test data.
- Accuracy of the cross-lingual transfer model (trained on English) on German test data.

---

## **Key Features**
- Demonstrates the power of fine-tuning for domain-specific tasks.
- Explores cross-lingual transfer learning between English and German.
- Simple and efficient workflow using AutoGluon for multimodal machine learning.

This notebook is ideal for understanding how fine-tuning and transfer learning can be applied to multilingual datasets for sentiment analysis.

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Sample data - replace with actual Cross-Lingual Amazon Product Review Sentiment dataset
# For demonstration, we create two simple DataFrames for English and German.
english_data = pd.DataFrame({
    'sentence': ['I love this product', 'This is terrible', 'Great quality', 'Not worth the money'],
    'label': [1, 0, 1, 0]  # 1 for positive, 0 for negative
})

german_data = pd.DataFrame({
    'sentence': ['Ich liebe dieses Produkt', 'Das ist schrecklich', 'Großartige Qualität', 'Nicht das Geld wert'],
    'label': [1, 0, 1, 0]  # 1 for positive, 0 for negative
})

# Split the German data for training and testing
train_data_german, test_data_german = train_test_split(german_data, test_size=0.2, random_state=42)

# Split the English data for cross-lingual transfer
train_data_english, test_data_english = train_test_split(english_data, test_size=0.2, random_state=42)
```

```python
from autogluon.multimodal import MultiModalPredictor

# Finetune the German BERT model
predictor_german = MultiModalPredictor(label='label', eval_metric='accuracy')

# Train on the German data
predictor_german.fit(train_data_german, time_limit=600)  # Train for 10 minutes

# Evaluate the model on the German test data
german_predictions = predictor_german.predict(test_data_german)
german_accuracy = predictor_german.evaluate(test_data_german)
print("German BERT Model Accuracy:", german_accuracy)
```

```python
# Train the model on the English data
predictor_english = MultiModalPredictor(label='label', eval_metric='accuracy')

# Fit the model on English data
predictor_english.fit(train_data_english, time_limit=600)  # Train for 10 minutes

# Evaluate the model on German test data (Cross-lingual transfer)
german_predictions_cross = predictor_english.predict(test_data_german)
german_accuracy_cross = predictor_english.evaluate(test_data_german)
print("Cross-lingual Transfer Model Accuracy:", german_accuracy_cross)
```

```python
print(f"Accuracy of the fine-tuned German BERT: {german_accuracy}")
print(f"Accuracy of the cross-lingual transfer (trained on English): {german_accuracy_cross}")
```
